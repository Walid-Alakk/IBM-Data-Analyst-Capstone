# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
# Query to fetch ConvertedComp data
QUERY = """
SELECT ConvertedComp
FROM master
"""
df = pd.read_sql_query(QUERY, conn)

# Plot histogram
plt.hist(df['ConvertedComp'].dropna(), bins=50, edgecolor='k', alpha=0.7)
plt.xlabel('Converted Compensation')
plt.ylabel('Frequency')
plt.title('Histogram of Converted Compensation')
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
# Query to fetch Age data
QUERY = """
SELECT Age
FROM master
"""
df = pd.read_sql_query(QUERY, conn)

# Plot box plot
plt.boxplot(df['Age'].dropna(), vert=False)
plt.xlabel('Age')
plt.title('Box Plot of Age')
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here

# Query to fetch Age and WorkWeekHrs data
QUERY = """
SELECT Age, WorkWeekHrs
FROM master
"""
df = pd.read_sql_query(QUERY, conn)

# Plot scatter plot
plt.scatter(df['Age'], df['WorkWeekHrs'], alpha=0.5)
plt.xlabel('Age')
plt.ylabel('Work Week Hours')
plt.title('Scatter Plot of Age and Work Week Hours')
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here

# Query to fetch Age, WorkWeekHrs, and CodeRevHrs data
QUERY = """
SELECT Age, WorkWeekHrs, CodeRevHrs
FROM master
"""
df = pd.read_sql_query(QUERY, conn)

# Plot bubble plot
plt.scatter(df['WorkWeekHrs'], df['CodeRevHrs'], s=df['Age'], alpha=0.5)
plt.xlabel('Work Week Hours')
plt.ylabel('Code Review Hours')
plt.title('Bubble Plot of Work Week Hours and Code Review Hours')
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here

# Query to fetch database data
QUERY = """
SELECT DatabaseDesireNextYear, COUNT(*) as count
FROM DatabaseDesireNextYear
GROUP BY DatabaseDesireNextYear
ORDER BY count DESC
LIMIT 5
"""
df = pd.read_sql_query(QUERY, conn)

# Plot pie chart
plt.pie(df['count'], labels=df['DatabaseDesireNextYear'], autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Databases Respondents Wish to Learn Next Year')
plt.axis('equal')  # Equal aspect ratio ensures the pie is drawn as a circle.
plt.show()



# Query to fetch the data about languages respondents wish to learn next year
QUERY = """
SELECT LanguageDesireNextYear, COUNT(*) as count
FROM LanguageDesireNextYear
GROUP BY LanguageDesireNextYear
ORDER BY count DESC
"""
df = pd.read_sql_query(QUERY, conn)
# Reset index to get the rank
df.reset_index(drop=True, inplace=True)
# Find the rank of Python
python_rank = df[df['LanguageDesireNextYear'] == 'Python'].index[0] + 1  # Adding 1 because index is 0-based
print(f"Rank of Python: {python_rank}")


# Query to fetch the count of respondents who currently work with SQL
QUERY = """
SELECT COUNT(*) as count
FROM LanguageWorkedWith
WHERE LanguageWorkedWith LIKE '%SQL%'
"""
df = pd.read_sql_query(QUERY, conn)
# Get the count from the DataFrame
sql_count = df.loc[0, 'count']
print(f"Number of respondents who currently work with SQL: {sql_count}")



# Query to fetch the count of respondents who currently work with only MySQL
QUERY = """
SELECT COUNT(*) as count
FROM LanguageWorkedWith
WHERE LanguageWorkedWith LIKE 'MySQL'
    OR LanguageWorkedWith LIKE 'MySQL;%' 
    OR LanguageWorkedWith LIKE '%;MySQL' 
    OR LanguageWorkedWith LIKE '%;MySQL;%'
"""
df = pd.read_sql_query(QUERY, conn)
# Get the count from the DataFrame
mysql_only_count = df.loc[0, 'count']
print(f"Number of respondents who currently work with MySQL only: {mysql_only_count}")
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here

# Query to fetch median WorkWeekHrs and CodeRevHrs for age group 30 to 35
QUERY = """
SELECT Age, WorkWeekHrs, CodeRevHrs
FROM master
WHERE Age BETWEEN 30 AND 35
"""
df = pd.read_sql_query(QUERY, conn)

# Calculate median values
median_work_week_hrs = df['WorkWeekHrs'].median()
median_code_rev_hrs = df['CodeRevHrs'].median()

# Plot stacked chart
labels = ['Work Week Hours', 'Code Review Hours']
values = [median_work_week_hrs, median_code_rev_hrs]

plt.bar(labels, values)
plt.ylabel('Hours')
plt.title('Median Work Week Hours and Code Review Hours for Age Group 30-35')
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here

# Query to fetch ConvertedComp and Age data
QUERY = """
SELECT Age, ConvertedComp
FROM master
WHERE Age BETWEEN 45 AND 60
"""
df = pd.read_sql_query(QUERY, conn)
# Calculate median ConvertedComp for each age
median_converted_comp = df.groupby('Age')['ConvertedComp'].median()
# Plot line chart
median_converted_comp.plot(kind='line')
plt.xlabel('Age')
plt.ylabel('Median Converted Compensation')
plt.title('Median Converted Compensation for Ages 45 to 60')
plt.show()


# Query to fetch ConvertedComp and Age data
QUERY = """
SELECT Age, ConvertedComp
FROM master
WHERE Age BETWEEN 25 AND 30
"""
df = pd.read_sql_query(QUERY, conn)
# Calculate median ConvertedComp for each age
median_converted_comp = df.groupby('Age')['ConvertedComp'].median()
# Plot line chart
median_converted_comp.plot(kind='line')
plt.xlabel('Age')
plt.ylabel('Median Converted Compensation')
plt.title('Median Converted Compensation for Ages 25 to 30')
plt.show()


# Query to fetch and count each DevType
QUERY = """
SELECT DevType, COUNT(*) as count
FROM DevType
GROUP BY DevType
ORDER BY count DESC
"""
df_roles = pd.read_sql_query(QUERY, conn)

# Display the top roles
print(df_roles.head())
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here

# Query to fetch MainBranch data
QUERY = """
SELECT MainBranch, COUNT(*) as count
FROM master
GROUP BY MainBranch
"""
df = pd.read_sql_query(QUERY, conn)

# Plot horizontal bar chart
df.plot(kind='barh', x='MainBranch', y='count', legend=False)
plt.xlabel('Count')
plt.ylabel('Main Branch')
plt.title('Main Branch Distribution')
plt.show()
```
